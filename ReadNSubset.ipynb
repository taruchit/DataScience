Difference between loc and iloc

```python
#Creating a sample dataframe
sample_df=pd.DataFrame({
    'gender':['M','F','M','M','F'],
    'grade':['A','A','B','B','A'],
    'marks':[22,21,12,14,20],
    'id':['A101','A102','A103','A104','A105']
})

sample_df.head()
```

```python
#Using loc[2:4]
sample_df.loc[2:4]
```

In the above result we observed loc takes labels as arguments.
Thus, 2:4 means it prints row label 2, label 3, label 4

In loc we have label based indexing

```python
#Using iloc[2:4]
sample_df.iloc[2:4]
```

In the above result we observed iloc takes index as arguments. Thus, 2:4 means it prints row of index 2 and index 3.

In iloc we have position based indexing

```python
#Accessing value from 2nd row 4th column
sample_df["id"].iloc[2:3]
```

```python
sample_df.iloc[1,3]
```
